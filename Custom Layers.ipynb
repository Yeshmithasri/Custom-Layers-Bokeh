```python
from bokeh.plotting import figure, show
from bokeh.models import Label, BoxAnnotation
from bokeh.layouts import column

p = figure(plot_width=400, plot_height=400)
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
p.scatter(x, y)

annotation = Label(x=3, y=7, text="Important Point", text_font_size="12pt", text_color="red")
p.add_layout(annotation)

overlay = BoxAnnotation(left=2.5, right=4.5, fill_alpha=0.2, fill_color="green")
p.add_layout(overlay)

show(column(p))

```

```python
from bokeh.plotting import figure, show
from bokeh.models import CustomJS, ColumnDataSource
from bokeh.layouts import layout
```

```python
source = ColumnDataSource(data={
    'x': [1, 2, 3, 4, 5],
    'y': [6, 7, 2, 4, 5],
})
```

```python
p = figure(title='Custom Layers Example',x_axis_label='x',y_axis_label='y',width=400,height=400)
```

```python

p.line('x', 'y', source=source, line_width=2, line_color='blue', legend_label='Line Layer')

```

```python
p.circle('x', 'y', source=source, size=10, fill_color='orange', line_color='hotpink', legend_label='Circle Layer')
```

```python
p.square('x', 'y', source=source, size=10, fill_color='blue', line_color='violet', legend_label='Square Layer')
```

```python
p.title.text = "Bokeh Custom Layers"
p.legend.location = "top_right"
p.legend.title = "Layers"
```

```python
show(p)
```
